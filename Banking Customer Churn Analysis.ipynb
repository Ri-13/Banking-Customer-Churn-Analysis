```python
order = ["Yes","No"]

churnedCount = customerChurn_data["Churned"].value_counts().reindex(order)
#Figure -> Axes -> Plot elements 
plt.figure(figsize=(10,5))
plt.subplot(1,2,1) # plt.subplot() already created the Axes
ax= sns.countplot(x=customerChurn_data["Churned"], hue= customerChurn_data["Churned"],order=order,palette="tab10",legend=False) # using existing Axes and returns it
plt.title("Customer Churned Distribution", fontweight="bold", size=15)
ax.bar_label(ax.containers[0],fontweight="black",size=15,label_type="edge")

plt.subplot(1,2,2)
plt.pie(churnedCount.values, labels=churnedCount.index, autopct="%1.1f%%", colors=sns.color_palette("tab10"),
        textprops={"fontsize":18,"fontweight":"black"},explode=[0,0.1])
plt.title("Customer Churned Distribution", fontweight="black", size=15)
plt.show()
```

```python
def countplot2(column):
    churn_count = pd.crosstab(
            customerChurn_data[column],
            customerChurn_data["Churned"],
            #normalize="index"
        ) 

    churn_count = churn_count[["Yes","No"]]
    
    churn_pct = churn_count.div(churn_count.sum(axis=1), axis=0) * 100

    ax = churn_count.plot(
        kind="bar",
        figsize=(10,5),
        colormap ="Paired"
    )

    for i, container in enumerate(ax.containers):
        for bar, pct in zip(container, churn_pct.iloc[:, i]):
            ax.annotate(
                f"{pct:.1f}%",
                (bar.get_x() + bar.get_width()/2, bar.get_height()),
                ha="center",
                va="bottom",
                fontweight="black",
                fontsize=12
            )

    ax.set_ylabel("Count")
    ax.set_xlabel(column)
    ax.set_title(f"Customer Churned by {column}", fontweight="black", size=15)
    ax.legend(title="Churned")
    #ax.set_ylim(0,100)

    plt.show()
```

```python
countplot2("Gender")
```

```python
def continuous_plot(column):
    plt.figure(figsize=(13,6))
    plt.subplot(1,2,1)
    sns.histplot(x=column,hue="Churned",data=customerChurn_data,kde=True,palette="tab10")
    plt.title(f"Distribution of {column} by Churn Status",fontweight="black",pad=20,size=15)

    plt.subplot(1,2,2)
    sns.boxplot(x="Churned",y=column,hue="Churned",data=customerChurn_data,palette="tab10",legend=False)
    plt.title(f"Distribution of {column} by Churn Status",fontweight="black",pad=20,size=15)
    plt.tight_layout()
    plt.show()
```

```python
continuous_plot("CreditScore")
```

```python
continuous_plot("Age")
```
